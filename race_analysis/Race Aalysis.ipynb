```python
import glob
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
data_path='./race_analysis'
```

```python
file_path=glob.glob(data_path+'/*.csv')
```

```python
diag_race=pd.read_csv(file_path[0])
```

```python
diag_race=diag_race.drop(columns=['Unnamed: 0'])
```

```python
mean_diag_race=diag_race.mean(axis=0)
```

```python
mean_diag_race[:10]
```

```python
census_race=pd.read_csv(file_path[1])
```

```python
columns_to_select=[
    'geography code',
    'Sex: All persons; Age: Age 65 and over; Ethnic Group: All categories: Ethnic group; measures: Value',
    'Sex: All persons; Age: Age 65 and over; Ethnic Group: White: Total; measures: Value',
    'Sex: All persons; Age: Age 65 and over; Ethnic Group: Mixed/multiple ethnic group: Total; measures: Value',
    'Sex: All persons; Age: Age 65 and over; Ethnic Group: Asian/Asian British: Total; measures: Value',
    'Sex: All persons; Age: Age 65 and over; Ethnic Group: Black/African/Caribbean/Black British: Total; measures: Value',
    'Sex: All persons; Age: Age 65 and over; Ethnic Group: Other ethnic group: Total; measures: Value'
]
```

```python
census_race_selected=census_race[columns_to_select]
```

```python
columns_to_swap=[
    'geography code',
    'Total',
    'White',
    'Mixed',
    'Asian',
    'Black',
    'Other'
]
census_race_selected.columns=columns_to_swap
```

```python
mean_diag_race=mean_diag_race.round(0)
```

```python
import numpy as np
```

```python
for entry in ['B','A','W','M','O','N']:
    census_race_selected[entry]=[np.nan]*census_race_selected.shape[0]
```

```python
for key,value in zip(mean_diag_race.index.tolist(),mean_diag_race.values.tolist()):
    geo_code,race_code=key.split('_')
    index_toadd=census_race_selected.index[census_race_selected['geography code'] == geo_code].tolist()
    if index_toadd:
        census_race_selected.at[index_toadd[0],race_code]=value
    
```

```python
import copy
```

```python
census_race_selected_copy_copy=copy.deepcopy(census_race_selected_copy)
```

```python
W_value_list=[]
B_value_list=[]
A_value_list=[]
O_value_list=[]
M_ratio_list=[]
for index,row in census_race_selected_copy.iterrows():
    W_ratio=row['White']/row['Total']
    B_ratio=row['Black']/row['Total']
    A_ratio=row['Asian']/row['Total']
    O_ratio=row['Other']/row['Total']
    M_ratio=row['Mixed']/row['Total']
    W_value=row['W']+row['N']*W_ratio
    W_value_list.append(W_value)
    B_value=row['N']*B_ratio
    B_value_list.append(B_value)
    A_value=row['N']*A_ratio
    A_value_list.append(A_value)
    O_value=row['N']*O_ratio
    O_value_list.append(O_value)
    M_value=row['N']*M_ratio
    M_ratio_list.append(M_value)
print('adjustment complete')
    
```

```python
census_race_selected_copy['W']=W_value_list
```

```python
census_race_selected_copy['B']=B_value_list
census_race_selected_copy['A']=A_value_list
census_race_selected_copy['O']=O_value_list
census_race_selected_copy['M']=M_ratio_list
```

```python
census_race_selected_copy=census_race_selected_copy.drop(columns=['N'])
```

```python
GeoCode=[]
Cencus_ratio_W=[]
Diagnosis_ratio_W=[]
Cencus_ratio_A=[]
Diagnosis_ratio_A=[]
Cencus_ratio_B=[]
Diagnosis_ratio_B=[]
Cencus_ratio_M=[]
Diagnosis_ratio_M=[]
Cencus_ratio_O=[]
Diagnosis_ratio_O=[]
for index,row in census_race_selected_copy.iterrows():
    total_diag=row['B']+row['A']+row['W']+row['M']+row['O']
    GeoCode.append(row['geography code'])
    Cencus_ratio_W.append(row['White']/row['Total'])
    Diagnosis_ratio_W.append(row['W']/total_diag)
    Cencus_ratio_A.append(row['Asian']/row['Total'])
    Diagnosis_ratio_A.append(row['A']/total_diag)
    Cencus_ratio_B.append(row['Black']/row['Total'])
    Diagnosis_ratio_B.append(row['B']/total_diag)
    Cencus_ratio_M.append(row['Mixed']/row['Total'])
    Diagnosis_ratio_M.append(row['M']/total_diag)
    Cencus_ratio_O.append(row['Other']/row['Total'])
    Diagnosis_ratio_O.append(row['O']/total_diag)
```

```python
GeoCode=[]
Cencus_ratio_W=[]
Diagnosis_ratio_W=[]
Diagnosis_W=[]
Cencus_ratio_A=[]
Diagnosis_ratio_A=[]
Diagnosis_A=[]
Cencus_ratio_B=[]
Diagnosis_ratio_B=[]
Diagnosis_B=[]
Cencus_ratio_M=[]
Diagnosis_ratio_M=[]
Diagnosis_M=[]
Cencus_ratio_O=[]
Diagnosis_ratio_O=[]
Diagnosis_O=[]
for index,row in census_race_selected_copy_copy.iterrows():
    total_diag=row['B']+row['A']+row['W']+row['M']+row['O']
    Diagnosis_W.append(row['W'])
    Diagnosis_A.append(row['A'])
    Diagnosis_B.append(row['B'])
    Diagnosis_M.append(row['M'])
    Diagnosis_O.append(row['O'])
    GeoCode.append(row['geography code'])
    Cencus_ratio_W.append(row['White']/row['Total'])
    Diagnosis_ratio_W.append(row['W']/total_diag)
    Cencus_ratio_A.append(row['Asian']/row['Total'])
    Diagnosis_ratio_A.append(row['A']/total_diag)
    Cencus_ratio_B.append(row['Black']/row['Total'])
    Diagnosis_ratio_B.append(row['B']/total_diag)
    Cencus_ratio_M.append(row['Mixed']/row['Total'])
    Diagnosis_ratio_M.append(row['M']/total_diag)
    Cencus_ratio_O.append(row['Other']/row['Total'])
    Diagnosis_ratio_O.append(row['O']/total_diag)
```

```python
W_comparison = pd.DataFrame(
    {'GeoCode': GeoCode,
     'Cencus_ratio_W': Cencus_ratio_W,
     'Diagnosis_ratio_W': Diagnosis_ratio_W,
     'Diagnosis_number':Diagnosis_M
    })
```

```python
W_comparison=W_comparison.dropna(axis=0)
```

```python
rel_dif=[]
for index,row in W_comparison.iterrows():
    rel_dif.append((row['Diagnosis_ratio_W']-row['Cencus_ratio_W'])/row['Cencus_ratio_W'])
```

```python
W_comparison['Relative_difference']=rel_dif
```

```python
B_comparison = pd.DataFrame(
    {'GeoCode': GeoCode,
     'Cencus_ratio_B': Cencus_ratio_B,
     'Diagnosis_ratio_B': Diagnosis_ratio_B,
     'Diagnosis_number':Diagnosis_B
    })
B_comparison=B_comparison.dropna(axis=0)
rel_dif=[]
for index,row in B_comparison.iterrows():
    rel_dif.append((row['Diagnosis_ratio_B']-row['Cencus_ratio_B'])/row['Cencus_ratio_B'])
B_comparison['Relative_difference']=rel_dif
```

```python
M_comparison = pd.DataFrame(
    {'GeoCode': GeoCode,
     'Cencus_ratio_M': Cencus_ratio_M,
     'Diagnosis_ratio_M': Diagnosis_ratio_M,
     'Diagnosis_number':Diagnosis_M
    })
M_comparison=M_comparison.dropna(axis=0)
rel_dif=[]
for index,row in M_comparison.iterrows():
    rel_dif.append((row['Diagnosis_ratio_M']-row['Cencus_ratio_M'])/row['Cencus_ratio_M'])
M_comparison['Relative_difference']=rel_dif
```

```python
O_comparison = pd.DataFrame(
    {'GeoCode': GeoCode,
     'Cencus_ratio_O': Cencus_ratio_O,
     'Diagnosis_ratio_O': Diagnosis_ratio_O,
     'Diagnosis_number':Diagnosis_O
    })
O_comparison=O_comparison.dropna(axis=0)
rel_dif=[]
for index,row in O_comparison.iterrows():
    rel_dif.append((row['Diagnosis_ratio_O']-row['Cencus_ratio_O'])/(row['Cencus_ratio_O']+1e-5))
O_comparison['Relative_difference']=rel_dif
```
