```python
column_for_finaldf=list(set(demdiag1['GEOGRAPHY_CODE'].tolist()))
column_for_race=[]
for column in column_for_finaldf:
    column_for_race.append(column+'_B')
    column_for_race.append(column+'_A')
    column_for_race.append(column+'_W')
    column_for_race.append(column+'_M')
    column_for_race.append(column+'_O')
    column_for_race.append(column+'_N')
```

```python
demdiag_race = pd.DataFrame(columns = column_for_race)
```

```python
for file in tqdm(file_path):
    row_name=file.split('demdiag')[-1].split('.')[0]
    demdiag=pd.read_csv(file)
    columns_name=['GEOGRAPHY_CODE','Measure','Value','COMMISSIONER_ORGANISATION_CODE']
    demdiag=demdiag[columns_name]
    for index,row in demdiag.iterrows():
        try:
            letters=list(row['Measure'])
            demdiag_race.loc[row_name,row['GEOGRAPHY_CODE']+'_'+letters[0]]=row['Value']
        except TypeError:
            try:
                geo_code=org_geo_dict[row['COMMISSIONER_ORGANISATION_CODE']]
                letters=list(row['Measure'])
                demdiag_race.loc[row_name,geo_code+'_'+letters[0]]=row['Value']
            except KeyError:
                continue
                
```

```python
demdiag_race
```

```python
demdiag_race.to_csv('demdiag_race.csv',index=True)
```
