```python
for file in file_path:
    row_name=file.split('pop')[-1].split('.')[0]
    pop_temp=pd.read_csv(file)
    select_index=pop_temp.columns.tolist().index('60')
    columns_to_sum=pop_temp.columns.tolist()[select_index:]
    pop_numbers=pop_temp[columns_to_sum]
    column_to_iter=pop_numbers.columns.tolist()
    for index,row in pop_numbers.iterrows():
        for column in column_to_iter:
            if isinstance(row[column], str):
                numbers=row[column].split(',')
                if len(numbers)==2:
                    row[column]=int(numbers[0])*1000+int(numbers[1])
                else:
                    row[column]=int(numbers[0])
    pop_temp['Sum']=pop_numbers.sum(axis=1)
    columns_to_choose=['Area Codes','Sum']
    pop_accounting=pop_temp[columns_to_choose]
    for index,row in pop_accounting.iterrows():
        pop_df.loc[row_name,row['Area Codes']]=row['Sum']
```
