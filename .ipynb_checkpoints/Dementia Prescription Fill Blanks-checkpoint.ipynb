```python
import glob
import pandas as pd
from tqdm import tqdm
```

```python
data_path='./time_series_data/Dementia'
```

```python
file_path=glob.glob(data_path+'/*.csv')
```

```python
file_path=sorted(file_path)
```

```python
og_df=pd.read_csv(file_path[0],index_col=0)
```

```python
for file in file_path[1:]:
    print(file)
    join_ondf=pd.read_csv(file)
    data_join=dict(zip(join_ondf['row_id'].tolist(),join_ondf['quantity'].tolist()))
    #extract date data
    number=file.split('/')[-1].split('.')[0]
    date=number[:4]+'-'+number[4:]+'-01'
    data_join_df=pd.DataFrame.from_dict(data_join,orient='index')
    data_join_df_t=data_join_df.transpose()
    date_series=pd.Series([date])
    data_join_df_t=data_join_df_t.set_index(date_series)
    og_df_test=og_df.append(data_join_df_t,ignore_index=False,verify_integrity=True)
    og_df=og_df_test
og_df_test=og_df_test.sort_index(axis=0)
```

```python
og_df_test.to_csv('dementia_updated.csv',index=True)
```
